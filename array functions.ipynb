```python

from pyspark.sql import functions as F

df = spark.createDataFrame([
    (1, [20, 10, 30]),
    (2, [40, 50])
], ["id", "nums"])

df.show(truncate=False)
```

```python

df.select(F.array(F.lit(1), F.lit(2), F.lit(3)).alias("arr")).show()
```

```python
#size-no of elements
df.select("id", F.size("nums").alias("length")).show()

#check if the value present or not
df.select("id", F.array_contains("nums", 20).alias("has_20")).show()

#explode-convert multiple values into multiple rows
df.select("id", F.explode("nums").alias("num")).show()

#accessin array element
df.select("id", df.nums[0].alias("first_num")).show()

#remove duplictes
df.select(F.array_distinct(F.array(F.lit(1), F.lit(1), F.lit(2)))).show()

#sorting array
df.select(F.array_sort("nums").alias("sorted")).show()

#adding elements
df.select(F.concat("nums", F.array(F.lit(100), F.lit(200))).alias("new_array")).show(truncate=False)

#extracting
df.select(F.slice("nums", 1, 2).alias("subset")).show()
```
