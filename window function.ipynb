```python
#window function
from pyspark.sql.window import Window
from pyspark.sql.functions import *
data = [
    ("IT", "Alice", 5000),
    ("IT", "Bob", 6000),
    ("HR", "Carol", 3500),
    ("HR","david",4000),
    ("HR", "Dave", 4000)
]

df = spark.createDataFrame(data, ["dept", "name", "salary"])
df.show()
```

```python
#row number
window = Window.partitionBy("dept").orderBy(col("salary").desc())

df.withColumn("row_num", row_number().over(window)).show()
```

```python
#rank
window = Window.partitionBy("dept").orderBy(col("salary").desc())

df.withColumn("rank", rank().over(window)).show()
```

```python
#dense rank
window = Window.partitionBy("dept").orderBy(col("salary").desc())

df.withColumn("dense_rank", dense_rank().over(window)).show()
```

```python

window = Window.partitionBy("dept").orderBy(col("salary").desc())

df.withColumn("prev_salary", lag("salary").over(window)) \
  .withColumn("next_salary", lead("salary").over(window)) \
  .show()
```
