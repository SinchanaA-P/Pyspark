```python
from pyspark.sql.functions import *

data = [("Alice Johnson", "  hello world  "),
        ("Bob", "Spark is FAST"),
        ("Charlie", None)]

df = spark.createDataFrame(data, ["name", "text"])
df.show()

#changing the alphabet case
df.select(lower(col("name")).alias("lower_name")).show()
df.select(upper(col("name")).alias("upper_name")).show()

#removing leading and trailing spaces
df.select(trim(col("text")).alias("trimmed")).show()
df.select(ltrim(col("text"))).show()
df.select(rtrim(col("text"))).show()

#counting the number of characters
df.select(length(col("name"))).show()


#substring-finding the substring in string

from pyspark.sql.functions import substring
df.select(substring("name", 1, 3)).show()   





```

```python
#splitting the string 
from pyspark.sql.functions import split
df.select(split("name"," ")).show()
df.select(split("name", "o")).show()
```

```python
#concate-join two string
from pyspark.sql.functions import concat
df.select(concat("name", "text")).show()
df.select(concat("name", "text")).show(truncate=False)
```

```python
#concat_ws-join two string with seperator
from pyspark.sql.functions import concat_ws
df.select(concat_ws(" ", "name", "text").alias("joined")).show(truncate=False)
df.select(concat_ws("-", "name", "text").alias("joined")).show(truncate=False)
```

```python
#convert the string "abc" to"xyz
from pyspark.sql.functions import translate
df.select(translate("name", "abc", "xyz")).show()
```

```python
#place of the letter in the string

df.select(instr("name", "a")).show()
df.select(instr("name", "i")).show()
```

```python
#extract portion mentioned in the string

df.select(regexp_extract(col("name"), "([a-zA-Z]+)", 1).alias("letters")).show()

df.select(regexp_extract(col("name"), "([a-z]+)", 1).alias("letters")).show()
```
