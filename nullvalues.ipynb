```python

df = spark.read.format("csv").option("header", "true") \
    .load("/databricks-datasets/retail-org/customers/customers.csv")
df.show(5)
```

```python
from pyspark.sql.functions import col
df.filter(col("tax_id").isNull()).count()

```

```python

df.filter(col("tax_id").isNotNull()).count()
```

```python

df.fillna({"city": "unknown", "tax_code": 1000}).show(10)
```

```python

df.fillna("Unknown").show(10)

```

```python

df.dropna().count()
```

```python
#replace particular value
df.na.replace("ADDRESS", "unknown").show(5)

#multiple values
df.na.replace(["ADDRESS", "Indiana"], ["VIENNA", "India"]).show(5)

```
