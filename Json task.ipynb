```python
#reading json folder
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, lit, struct
df_raw = spark.read \
    .option("multiline", "true") \
    .json("/Volumes/workspace/default/volume")
df_raw.printSchema()


```

```python


df_flat = df_raw.select(
    col("Body.mnsn").alias("mnsn"),
    col("Body.timestamp").alias("createTimestamp"),
    col("Body.type").alias("type"),

    col("Properties.relatedGroupId").alias("relatedGroupId"),
    col("Properties.customerId").alias("customerId"),
    col("Properties.dealerId").alias("dealerId"),

    col("Body.smsProperties.device.serialId").alias("serialNumber"),
    col("Body.smsProperties.device.modelName").alias("modelName"),
    col("Body.smsProperties.device.familyName").alias("familyName"),
    col("Body.smsProperties.device.friendlyName").alias("friendlyName"),
    col("Body.smsProperties.device.statusRawValue").alias("statusCode"),
    col("Body.smsProperties.device.address").alias("location"),

    col("Body.smsProperties.interface.ethernetList").alias("ethernetList"),
    col("Body.smsProperties.interface.ipList").alias("ipList"),

    col("Body.smsProperties.device.inTrayList").alias("inTrayList"),
    col("Body.smsProperties.device.outTrayList").alias("outTrayList"),
    col("Body.smsProperties.device.markerList").alias("markerList"),
    col("Body.smsProperties.device.disposalMarkerList").alias("disposalMarkerList"),

    col("Body.smsProperties.device.suppliesCounter.TYPE.toner.large.end.black.value").alias("black"),
    col("Body.smsProperties.device.suppliesCounter.TYPE.toner.large.end.cyan.value").alias("cyan"),
    col("Body.smsProperties.device.suppliesCounter.TYPE.toner.large.end.magenta.value").alias("magenta"),
    col("Body.smsProperties.device.suppliesCounter.TYPE.toner.large.end.yellow.value").alias("yellow")
)
```

```python
df_final = df_flat.select(
    "mnsn",
    "createTimestamp",
    "type",
    "relatedGroupId",
    "customerId",
    "dealerId",

    struct(
        "serialNumber",
        "modelName",
        "familyName",
        "friendlyName",
        "statusCode",
        "location"
    ).alias("deviceGeneral"),

    struct(
        "ethernetList",
        "ipList"
    ).alias("interface"),

    "inTrayList",
    "outTrayList",
    "markerList",
    "disposalMarkerList",

    
)
```

```python
df_final.printSchema()
df_final.display()
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import col, row_number

# Define a window partitioned by mnsn and ordered by createTimestamp descending
window_spec = Window.partitionBy("mnsn").orderBy(col("createTimestamp").desc())

# Add row number and keep only the first (latest) row per mnsn
df_clean = df_final.withColumn("row_num", row_number().over(window_spec)) \
             .filter(col("row_num") == 1) \
             .drop("row_num")
df_clean.display()
```

```python
#creating table according to schema of dataframe
spark.sql("CREATE DATABASE IF NOT EXISTS iot")

(
    df_clean
    .write
    .format("delta")
    .mode("overwrite")              # first load only
    .option("overwriteSchema", "true")
    .saveAsTable("iot.device_latest")
)
```
